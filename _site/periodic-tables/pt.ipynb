```python
from pymatgen.util.plotting import periodic_table_heatmap
import pandas as pd
from mendeleev import element
```

```python
path1 = "/Users/robertwexler/Dropbox (Princeton)/ML-for-TWS/ml/data/reference/Cohesive-energies"
path1 += "/element_cohesive_energies.csv"
df1 = pd.read_csv(path1)
df1["Element"] = df1["Element"].apply(lambda x: element(x).symbol)
```

```python
ed = {}
for ind, row in df1.iterrows():
    ed[row.Element] = row["Per Atom"]
```

```python
periodic_table_heatmap(elemental_data=ed, value_format="%.2f")
```
